#2025Q2 Revenue Calculation

> This query calculates key metrics including net revenue, order count, and average order value for Q2 of 2024 and 2025 from the `order_items` table. It filters for orders with 'Complete', 'Shipped', or 'Processing' statuses within the specified quarters and years. The results are presented by year and quarter, providing a comparative view of performance between the two periods.



```python
%%bigquery --project {project_id}
select
  FORMAT_DATETIME('%Y-%QQ', created_at) as year_quarter,
  round(sum(sale_price),2)  as net_revenue,
  count(distinct(order_id)) as count_order_id,
  round(round(sum(sale_price),2) / count(distinct(order_id)),2) as average_order_value
  from
  bigquery-public-data.thelook_ecommerce.order_items
where
  extract(QUARTER FROM created_at) = 2 and
  extract(YEAR FROM created_at) = 2025 and
  status IN ('Complete','Shipped', 'Processing')
group by
  year_quarter

UNION ALL
select
  FORMAT_DATETIME('%Y-%QQ', created_at) as year_quarter,
  round(sum(sale_price),2)  as net_revenue,
  count(distinct(order_id)) as count_order_id,
  round(round(sum(sale_price),2) / count(distinct(order_id)),2) as average_order_value
from
  bigquery-public-data.thelook_ecommerce.order_items
where
  extract(QUARTER FROM created_at) = 2 and
  extract(YEAR FROM created_at) = 2024 and
  status IN ('Complete','Shipped', 'Processing')
group by
  year_quarter
```

```python
%%bigquery --project {project_id}
select
  FORMAT_DATETIME('%Y-%QQ', created_at) as year_quarter,
  status as order_status,
  round(sum(sale_price),2)  as net_revenue,
  from
  bigquery-public-data.thelook_ecommerce.order_items
where
  extract(QUARTER FROM created_at) = 2 and
  extract(YEAR FROM created_at) = 2025 and
  status IN ('Complete','Shipped', 'Processing')
group by
  year_quarter,
  status
```
